```python
class Video(object):
    
    def __init__(self, video_id, captions=[], sentence_ids=[]):
        self.video_id = video_id
        self.captions = captions
        self.sentence_ids = sentence_ids
    
    def __str__(self):
        return '{}'.format(self.video_id)
```

## Converting videos into frames

```python
import cv2
import os
from pathlib import Path
```

```python
def getFrame(sec, dir_name):

    cap.set(cv2.CAP_PROP_POS_MSEC, sec*1000)
    hasFrames, image = cap.read()
    if hasFrames:
        cv2.imwrite("./frames/" + dir_name + "/frame"+ str(count)+ ".jpg", image)
    return hasFrames

```

```python
for i, video_name in enumerate(os.listdir('./train-video')):
    
    dir_name = video_name.split('.')[0]
    Path('./frames/' + dir_name).mkdir(parents=True, exist_ok=True)
    
    print(video_name)
    cap = cv2.VideoCapture('./train-video/'+ video_name)

    sec = 0
    frameRate = 1/3 
    count = 1
    success = getFrame(sec, dir_name)
    while success:
        
        count += 1
        sec = sec + frameRate
        sec = round(sec, 2)
        if count == 33:
            success = False
        else:
            success = getFrame(sec, dir_name)
```
